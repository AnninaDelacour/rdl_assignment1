In dieser Aufgabe geht es darum mittels eines Modells einen Text aus dem Excerpt von dem Paper "Attention is all you need" einen weiteren Text zu generieren. Ich persöhnlich habe festgestellt, dass es schwierig ist ein Modell zu finden, dass in Google Colab funktioniert und auch gute Ergebnisse liefert. Am Ende habe ihc mich für das Modell "facebook/opt-1.3b" von Huggingface entschieden, da dies zumindest in einem Setting annäherned brauchbare Ergebnisse findet. Jedoch ist die Länge der Prompts mit dem Excerpt aus dem Paper stark limitiert, da bei einem zu langen Prompt der RAM von Google Colab free überschritten wird und das Notebook crasht. Das Modell hat zwei Ausführungen für einmal Temperature, k und p, wobei bei Modell 2 die Parameter häher sind und auch der Output meistens nicht sinnvoll nutzbar. Daran ändert auch die Veränderung der maximalen Länge des Outputs nicht

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load a pre-trained text generation model from Hugging Face
model_name = "facebook/opt-1.3b"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Function to generate text
def generate_text(prompt, max_length=100, temperature=1.0, top_k=50, top_p=0.95, repetition_penalty=1.0):
    input_ids = tokenizer.encode(prompt, return_tensors="pt")
    output = model.generate(
        input_ids,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        repetition_penalty=repetition_penalty,
        do_sample=True
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Example prompt from research context
research_prompt =  """Tell me something about Multi head attention and make it as simple as possible"""

# Generate text with different parameters
output_1 = generate_text(research_prompt, max_length=150, temperature=0.7, top_k=40, top_p=0.9, repetition_penalty=1.2)
output_2 = generate_text(research_prompt, max_length=150, temperature=1.2, top_k=100, top_p=0.98, repetition_penalty=1.0)

# Print results
print("\n=== Text with Controlled Creativity (Predictable) ===\n")
print(output_1)
print("\n=== Text with High Creativity (More Random) ===\n")
print(output_2)

# Analysis Section
def analyze_results(original_text, generated_texts):
    print("\n=== Analysis of Generated Text ===\n")
    print("Original Text:")
    print(original_text)


# Run analysis
analyze_results(research_prompt, [output_1, output_2])
```
